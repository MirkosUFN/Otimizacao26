```python
import matplotlib.pyplot as plt
import numpy as np

def calcular_e_plotar_reta():
    print("--- Gerador de Equação da Reta ---")
    try:
        # Entrada de dados para o primeiro ponto
        x1 = float(input("Digite x1: "))
        y1 = float(input("Digite y1: "))

        # Entrada de dados para o segundo ponto
        x2 = float(input("Digite x2: "))
        y2 = float(input("Digite y2: "))

        if x1 == x2:
            if y1 == y2:
                print("\nErro: Os pontos são idênticos, não é possível definir uma reta única.")
                return
            else:
                print(f"\nReta Vertical: x = {x1}")
                plotar_grafico(x1, y1, x2, y2, vertical=True)
                return

        # Cálculo do coeficiente angular (a) e linear (b)
        a = (y2 - y1) / (x2 - x1)
        b = y1 - a * x1

        # Exibição da equação
        sinal = "+" if b >= 0 else "-"
        print(f"\nEquação da reta: y = {a:.2f}x {sinal} {abs(b):.2f}")

        plotar_grafico(x1, y1, x2, y2, a, b)

    except ValueError:
        print("Erro: Por favor, insira apenas valores numéricos.")

def plotar_grafico(x1, y1, x2, y2, a=None, b=None, vertical=False):
    plt.figure(figsize=(8, 6))

    # Plota os dois pontos informados
    plt.scatter([x1, x2], [y1, y2], color='red', zorder=5, label='Pontos Informados')

    if vertical:
        plt.axvline(x=x1, color='blue', label=f'x = {x1}')
        y_min, y_max = min(y1, y2) - 5, max(y1, y2) + 5
        plt.ylim(y_min, y_max)
    else:
        # Cria um intervalo de X para desenhar a linha além dos pontos
        margem = abs(x2 - x1) * 0.5 if x1 != x2 else 5
        x_range = np.linspace(min(x1, x2) - margem, max(x1, x2) + margem, 100)
        y_range = a * x_range + b
        plt.plot(x_range, y_range, color='blue', label='Reta y = ax + b')

    # Configurações do plano cartesiano
    plt.axhline(0, color='black', linewidth=1)
    plt.axvline(0, color='black', linewidth=1)
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.legend()
    plt.title("Gráfico da Reta")
    plt.xlabel("Eixo X")
    plt.ylabel("Eixo Y")

    plt.show()

if __name__ == "__main__":
    calcular_e_plotar_reta()
```
