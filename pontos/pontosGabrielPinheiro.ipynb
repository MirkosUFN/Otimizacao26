```python
import matplotlib.pyplot as plt


def plotar_pontos():
    x_coords = []
    y_coords = []

    print("--- Traçador de Pontos ---")

    while True:
        try:
            x = float(input("Digite a coordenada X do ponto: "))
            y = float(input("Digite a coordenada Y do ponto: "))

            x_coords.append(x)
            y_coords.append(y)

            continuar = (
                input("Quer adicionar outro ponto? (s/n): ").strip().lower()
            )
            if continuar != "s":
                break
        except ValueError:
            print("Por favor, digite um número válido.")

    # Criando o gráfico
    plt.figure(figsize=(8, 6))

    # Plota os pontos
    plt.scatter(
        x_coords,
        y_coords,
        color="red",
        s=100,
        zorder=3,
        label="Pontos inseridos",
    )

    # Adiciona rótulos com as coordenadas ao lado de cada ponto
    for x, y in zip(x_coords, y_coords):
        plt.annotate(
            f" ({x}, {y})",
            (x, y),
            textcoords="offset points",
            xytext=(5, 5),
            ha="left",
        )

    # Desenha os eixos cartesianos passando pela origem (0,0)
    plt.axhline(0, color="black", linewidth=1.2, linestyle="-")
    plt.axvline(0, color="black", linewidth=1.2, linestyle="-")

    # Configurações visuais do gráfico
    plt.grid(True, linestyle="--", alpha=0.6)
    plt.xlabel("Eixo X")
    plt.ylabel("Eixo Y")
    plt.title("Visualização de Pontos no Plano Cartesiano")
    plt.legend()

    # Mostra o gráfico na tela
    plt.show()


# Executa a função
if __name__ == "__main__":
    plotar_pontos()
```
