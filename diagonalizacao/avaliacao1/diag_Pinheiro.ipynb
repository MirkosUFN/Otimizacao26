```python

def imprimir_matriz(matriz, titulo=""):
    if titulo:
        print(f"\n  {titulo}")
    n = len(matriz)
    print("  " + "-" * (12 * (len(matriz[0]))))
    for linha in matriz:
        # Formatação para alinhar colunas e separar o termo independente
        valores = "  | "
        for j, v in enumerate(linha):
            if j == len(linha) - 1:
                valores += " | "
            valores += f"{v:10.2f} "
        valores += "|"
        print(valores)
    print("  " + "-" * (12 * (len(matriz[0]))))

def resolver_por_colunas(matriz):
    n = len(matriz)

    print("\n" + "=" * 65)
    print("  INICIANDO DIAGONALIZAÇÃO POR COLUNAS")
    print("=" * 65)
    imprimir_matriz(matriz, "Matriz Original")

    # O loop externo percorre cada COLUNA do sistema (cada variável)
    for j in range(n):
        print(f"\n>>> PROCESSANDO COLUNA {j + 1} <<<")

        # 1. Identificar o pivô na coluna atual (elemento da diagonal)
        pivo = matriz[j][j]

        # Tratamento de Pivô Zero (Troca de linhas para salvar a coluna)
        if abs(pivo) < 1e-10:
            for k in range(j + 1, n):
                if abs(matriz[k][j]) > 1e-10:
                    matriz[j], matriz[k] = matriz[k], matriz[j]
                    imprimir_matriz(matriz, f"Pivô nulo! Trocando Linha {j+1} com Linha {k+1}")
                    pivo = matriz[j][j]
                    break
            else:
                print(f"ERRO: A coluna {j+1} não possui pivô válido. Sistema sem solução única.")
                return None

        # 2. Normalizar a linha do pivô (tornar o elemento da diagonal = 1)
        if pivo != 1:
            print(f"  -> Normalizando Linha {j+1}: Dividindo por {pivo:.2f}")
            matriz[j] = [elemento / pivo for elemento in matriz[j]]
            imprimir_matriz(matriz, f"Coluna {j+1}: Pivô em (L{j+1}, C{j+1}) agora é 1")

        # 3. "Gerar os zeros" na coluna atual
        # Percorre todas as linhas para zerar os elementos ACIMA e ABAIXO do pivô
        print(f"  -> Eliminando valores da Coluna {j+1} nas outras linhas:")
        operou = False
        for i in range(n):
            if i != j: # Não processa a linha do próprio pivô
                fator = matriz[i][j]
                if abs(fator) > 1e-10: # Só opera se o valor não for zero
                    # L_i = L_i - fator * L_j (operação elementar)
                    matriz[i] = [L_i - (fator * L_j) for L_i, L_j in zip(matriz[i], matriz[j])]
                    print(f"     L{i+1} = L{i+1} - ({fator:.2f}) * L{j+1}")
                    operou = True

        if operou:
            imprimir_matriz(matriz, f"Resultado: Coluna {j+1} limpa (apenas pivô 1)")
        else:
            print(f"     A coluna {j+1} já estava limpa.")

    # Exibição Final
    print("\n" + "=" * 65)
    print("  SISTEMA RESOLVIDO")
    print("=" * 65)
    for i in range(n):
        print(f"  x{i+1} = {matriz[i][-1]:.2f}")
    print("=" * 65)

def iniciar():
    try:
        n = int(input("Quantas variáveis o sistema possui? "))
        matriz = []
        print(f"Digite os {n+1} valores de cada linha (coeficientes + resultado):")
        for i in range(n):
            linha = list(map(float, input(f"Linha {i+1}: ").split()))
            if len(linha) != n + 1:
                print("Quantidade de valores incorreta!")
                return
            matriz.append(linha)

        resolver_por_colunas(matriz)
    except ValueError:
        print("Entrada inválida. Use apenas números.")

if __name__ == "__main__":
    iniciar()
```
