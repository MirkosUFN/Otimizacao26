```python
Exercício 1: Otimização de Infraestrutura em Nuvem (IaaS)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Intervalo de x
x = np.linspace(0, 15, 400)

# Restrições
r1 = (10 - x) / 2      # x1 + 2x2 = 10
r2 = (12 - x) / 4      # x1 + 4x2 = 12
r3 = 12 - x            # x1 + x2 = 12

# Plot
plt.figure(figsize=(8,6))

plt.plot(x, r1, label='x1 + 2x2 = 10')
plt.plot(x, r2, label='x1 + 4x2 = 12')
plt.plot(x, r3, label='x1 + x2 = 12')

# Região viável aproximada
y_min = np.maximum(r1, r2)
y_max = r3

plt.fill_between(x, y_min, y_max,
                 where=(y_max >= y_min),
                 color='lightblue',
                 alpha=0.5)

# Pontos candidatos
points = [(8,1), (12,0)]

for p in points:
    plt.plot(p[0], p[1], 'ro')
    plt.text(p[0]+0.2, p[1], f'{p}')

# Configuração do gráfico
plt.xlim(0,15)
plt.ylim(0,10)

plt.xlabel('x1 = Instâncias Standard')
plt.ylabel('x2 = Instâncias High-Performance')

plt.title('Região Viável - Problema de PL')
plt.grid(True)
plt.legend()

plt.show()

# Solução ótima
x1 = 8
x2 = 1
Z = 20*x1 + 50*x2

print(f'Solução ótima:')
print(f'x1 = {x1}')
print(f'x2 = {x2}')
print(f'Custo mínimo = ${Z}/h')
```

```python
Exercício 2 — Alocação de Banda e Roteamento de Tráfego
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Intervalo de x
x = np.linspace(0, 10, 400)

# Restrições
r1 = (12 - x) / 2          # x1 + 2x2 = 12
r2 = (22 - 3*x) / 2       # 3x1 + 2x2 = 22

# Plot das retas
plt.figure(figsize=(8,6))

plt.plot(x, r1, label='x1 + 2x2 = 12')
plt.plot(x, r2, label='3x1 + 2x2 = 22')

# Região viável
y_max = np.minimum(r1, r2)

plt.fill_between(x, 0, y_max,
                 where=(y_max >= 0),
                 color='lightblue',
                 alpha=0.5)

# Vértices
points = [(0,0), (0,6), (5,3.5), (22/3,0)]

for p in points:
    plt.plot(p[0], p[1], 'ro')
    plt.text(p[0]+0.1, p[1]+0.1, f'{p}')

# Configurações
plt.xlim(0,10)
plt.ylim(0,12)

plt.xlabel('x1 = Stream Básico')
plt.ylabel('x2 = Stream Premium')

plt.title('Região Viável - Problema de PL')
plt.grid(True)
plt.legend()

plt.show()

# Solução ótima
x1 = 5
x2 = 3.5
Z = 3*x1 + 5*x2

print('Solução ótima:')
print(f'x1 = {x1}')
print(f'x2 = {x2}')
print(f'QoS máximo = {Z}')
```
