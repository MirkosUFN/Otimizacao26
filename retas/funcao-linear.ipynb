- como generalizar essa função;
- e como representar essa generalização por meio de um algoritmo.

# 2. Representação da função

Uma função de primeiro grau é definida por:

f(x) = ax + b

Onde:
- a representa a inclinação da reta;
- b representa o valor onde a reta intercepta o eixo y.

# 3. Interpretação no gráfico

O gráfico de uma função de primeiro grau é sempre uma reta.

Se a > 0, a reta é crescente.  
Se a < 0, a reta é decrescente.

# 4. Generalização

A generalização consiste em permitir que os valores de a e b variem.

Assim, a mesma expressão f(x) = ax + b representa uma família de retas.

```python
import numpy as np
import matplotlib.pyplot as plt

def desenhar_reta(a, b):
    # Cria valores de x no intervalo de -1 até 1
    x = np.linspace(-1, 1, 2)

    # Calcula os valores de y usando a função de primeiro grau
    y = a * x + b

    # Plota a reta
    plt.plot(x, y, label=f"f(x) = {a}x + {b}")

    # Desenha os eixos do plano cartesiano
    plt.axhline(0, color='black', linewidth=1)
    plt.axvline(0, color='black', linewidth=1)

    # Configurações do gráfico
    plt.title("Gráfico da função de primeiro grau")
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.grid(True)
    plt.legend()

    # Exibe o gráfico
    plt.show()

desenhar_reta(2, 1)
```
