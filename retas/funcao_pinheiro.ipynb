```python
import matplotlib.pyplot as plt
import numpy as np

def calcular_funcao():
    print("--- Calculadora da Função y = 5x - 11 ---")

    try:
        # 1. Entrada do usuário para um ponto específico
        valor_x = float(input("Digite o valor de x para calcular y: "))
        valor_y = 5 * valor_x - 11
        print(f"\nPara x = {valor_x}, o resultado de y é: {valor_y}")

        # 2. Preparação dos dados para o gráfico
        # Criamos um intervalo de x de -10 a 10 para desenhar a reta
        x_grafico = np.linspace(valor_x - 10, valor_x + 10, 400)
        y_grafico = 5 * x_grafico - 11

        # 3. Construção do Gráfico
        plt.figure(figsize=(8, 6))
        plt.plot(x_grafico, y_grafico, label='y = 5x - 11', color='blue', linewidth=2)

        # Destacando o ponto que você calculou
        plt.scatter(valor_x, valor_y, color='red', zorder=5, label=f'Ponto ({valor_x}, {valor_y})')

        # Configurações visuais
        plt.axhline(0, color='black',linewidth=1) # Eixo X
        plt.axvline(0, color='black',linewidth=1) # Eixo Y
        plt.title('Gráfico da Função de 1º Grau')
        plt.xlabel('Eixo X')
        plt.ylabel('Eixo Y')
        plt.grid(True, linestyle='--', alpha=0.7)
        plt.legend()

        print("\nExibindo o gráfico...")
        plt.show()

    except ValueError:
        print("Erro: Por favor, digite apenas números.")

if __name__ == "__main__":
    calcular_funcao()
```
