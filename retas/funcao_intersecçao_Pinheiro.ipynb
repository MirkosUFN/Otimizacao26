```python
import matplotlib.pyplot as plt
import numpy as np

def analisar_interseccao():
    print("--- Analisador de Intersecção de Duas Funções ---")

    # Entrada de dados para a Função 1: y = a1*x + b1
    a1 = float(input("Função 1 - Digite o coeficiente 'a' (ex: 5): "))
    b1 = float(input("Função 1 - Digite o coeficiente 'b' (ex: -11): "))

    # Entrada de dados para a Função 2: y = a2*x + b2
    a2 = float(input("\nFunção 2 - Digite o coeficiente 'a': "))
    b2 = float(input("Função 2 - Digite o coeficiente 'b': "))

    # Cálculo da intersecção
    interseccao_x = None
    if a1 == a2:
        if b1 == b2:
            resultado_texto = "As retas são idênticas (infinitos pontos)."
        else:
            resultado_texto = "As retas são paralelas e não se tocam."
    else:
        interseccao_x = (b2 - b1) / (a1 - a2)
        interseccao_y = a1 * interseccao_x + b1
        resultado_texto = f"As retas se cruzam no ponto: ({interseccao_x:.2f}, {interseccao_y:.2f})"

    print(f"\nResultado: {resultado_texto}")

    # Configuração do gráfico
    # Definimos o range de X baseado no ponto de encontro (se existir)
    base_x = interseccao_x if interseccao_x is not None else 0
    x = np.linspace(base_x - 10, base_x + 10, 400)

    y1 = a1 * x + b1
    y2 = a2 * x + b2

    plt.figure(figsize=(10, 6))
    plt.plot(x, y1, label=f'y = {a1}x + ({b1})', color='blue')
    plt.plot(x, y2, label=f'y = {a2}x + ({b2})', color='green', linestyle='--')

    # Se houver intersecção, desenha o ponto
    if interseccao_x is not None:
        plt.scatter(interseccao_x, interseccao_y, color='red', s=100, zorder=5, label='Intersecção')
        plt.annotate(f'({interseccao_x:.1f}, {interseccao_y:.1f})',
                     xy=(interseccao_x, interseccao_y), xytext=(interseccao_x+1, interseccao_y+2),
                     arrowprops=dict(facecolor='black', shrink=0.05, width=1, headwidth=5))

    plt.axhline(0, color='black', linewidth=1, alpha=0.3)
    plt.axvline(0, color='black', linewidth=1, alpha=0.3)
    plt.title('Intersecção de Funções de 1º Grau')
    plt.grid(True, alpha=0.3)
    plt.legend()
    plt.show()

if __name__ == "__main__":
    analisar_interseccao()
```
