```python
def calcular_funcao(a, b, x):
    """
    Calcula o valor de y para uma função de primeiro grau: y = ax + b
    """
    y = a * x + b
    return y

def executor():
    print("--- Calculadora de Função de Primeiro Grau (y = ax + b) ---")

    try:
        # Definindo os parâmetros da função
        valor_a = float(input("Digite o valor de 'a' (coeficiente angular): "))
        valor_b = float(input("Digite o valor de 'b' (termo constante): "))

        print(f"\nFunção definida: y = {valor_a}x + {valor_b}")
        print("Digite 'sair' para encerrar o programa.\n")

        while True:
            entrada = input("Digite o valor de x: ")

            if entrada.lower() == 'sair':
                print("Encerrando o programa...")
                break

            # Converte a entrada para float e calcula
            valor_x = float(entrada)
            resultado_y = calcular_funcao(valor_a, valor_b, valor_x)

            print(f"Para x = {valor_x}, o valor de y é: {resultado_y}")
            print("-" * 30)

    except ValueError:
        print("Erro: Por favor, insira apenas números válidos.")

# Inicia o programa
if __name__ == "__main__":
    executor()
```

```python
import random
import matplotlib.pyplot as plt

def funcha(a, b, x):
    """
    Retorna o valor de y para uma função de primeiro grau.
    """
    return a * x + b

def gerar_e_plotar_pentagrama_deformado():
    # Valores fixos de a e b
    a = 2
    b = 4

    # Listas para armazenar as coordenadas completas
    todos_x = []
    todos_y = []

    # 1. Geração dos 5 pontos aleatórios
    print(f"Gerando 5 pontos aleatórios para a={a} e b={b}:")
    for i in range(5):
        # Gera um valor aleatório para x entre 1 e 100
        x_rand = random.randint(1, 100)
        # Chama a função funcha
        y_res = funcha(a, b, x_rand)

        todos_x.append(x_rand)
        todos_y.append(y_res)
        print(f"Ponto {i}: (x={x_rand}, y={y_res})")

    # 2. Ordenar os pontos gerados aleatoriamente pelo valor de X.
    # Isso é necessário para que possamos nos referir a eles como Ponto 0 (mais à esquerda)
    # até Ponto 4 (mais à direita) de forma consistente.
    pontos = sorted(zip(todos_x, todos_y))
    x_ordenado, y_ordenado = zip(*pontos)

    # 3. Definir a ordem de conexão do "Pentagrama".
    # A ordem clássica de desenho de uma estrela de 5 pontas, pulando um ponto, é:
    # Início -> Ponto 0 -> Ponto 2 -> Ponto 4 -> Ponto 1 -> Ponto 3 -> Volta para 0.
    ordem_conexoes = [0, 2, 4, 1, 3, 0]

    # Prepara as listas de coordenadas X e Y na ordem correta do desenho
    x_desenho = [x_ordenado[i] for i in ordem_conexoes]
    y_desenho = [y_ordenado[i] for i in ordem_conexoes]

    # 4. Configurar e Plotar o Gráfico
    plt.figure(figsize=(10, 6))

    # Plota as linhas que "tentam" formar o pentagrama.
    # Como todos os pontos estão na mesma reta, as linhas vão e voltam sobre a mesma trajetória.
    plt.plot(x_desenho, y_desenho, color='blue', linewidth=2.5, label='Traçado do "Pentagrama" (Colinear)')

    # Plota os 5 pontos gerados individualmente para visualização
    plt.scatter(x_ordenado, y_ordenado, color='red', s=120, zorder=5, label='Pontos Gerados (y=2x+4)')

    # Ajustes estéticos e anotações
    plt.title(f'Pentagrama Deformado: Pontos Aleatórios de $y = {a}x + {b}$')
    plt.xlabel('Eixo X (Valores Aleatórios)')
    plt.ylabel('Eixo Y (f(x))')
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.legend()

    # Mostra o gráfico
    print("\nExibindo o gráfico...")
    print("Nota: O resultado visual é uma linha reta única porque todos os pontos são colineares.")
    plt.show()

if __name__ == "__main__":
        gerar_e_plotar_pentagrama_deformado()
```
